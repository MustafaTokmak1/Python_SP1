```python
import pandas as pd

#1. Hvor mange monumenter er der i København?

data_frame = pd.read_csv("https://wfs-kbhkort.kk.dk/k101/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=k101:monumenter&outputFormat=csv&SRSNAME=EPSG:4326")
amount_monuments = data_frame.shape[0]
print("Exercise 1 output:")
print("Copenhagen has " + str(amount_monuments) + " monuments")

```

```python
#2. Hvor mange monumenter bliver vedligeholdt? Dvs. graffitirenhold = ja

amount_maintained = data_frame.loc[data_frame['graffitirenhold']=='Ja']
print("Exercise 2 output:")
print(str(amount_maintained.shape[0]) + " monuments are maintained")
```

```python
import folium

#3. Lav en funktion som kan finde koordinaterne på et monument baseret på monumentets id eller navn?
#    F.eks:
#        def monumentById(monumentId):
#            return coordinates

def monument_by_id(monumentId):
    monument = data_frame.loc[data_frame['id']==monumentId]
    longtitude = monument.iloc[0,7]
    latitude = monument.iloc[0,8]
    return(latitude, longtitude)
    
    #example: monumenter.2	49865	H.W. Meyer	12.56281	55.696741	Ja
print("Exercise 3 output:")
print(monument_by_id(monumentId=49865))


#3.a Vis monument som bliver returneret i metoden, på et kort over københavn ved brug af plotting. (Se afsnittet om 'Folium and Bokeh' under notebooks/03-3 Plotting)

map_example = folium.Map(location=(monument_by_id(monumentId=49865)), zoom_start=17) #show map
folium.Marker(monument_by_id(monumentId=49865)).add_to(map_example) #add marker
print("\n Exercise 3a output:")
map_example
```
